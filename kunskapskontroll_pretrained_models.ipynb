```python
model = ResNet50(weights='imagenet')
```

```python
img_path = 'tennisboll.jpg' 
img = image.load_img(img_path, target_size=(224,224))
img
```

```python
x = image.img_to_array(img)
x.shape
```

```python
x = np.expand_dims(x, axis=0)
x.shape
```

```python
x = preprocess_input(x)
```

```python
preds = model.predict(x)
```

```python
print('Predicted:', decode_predictions(preds, top=1)[0])
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
```

```python
# Ladda VGG16 modell
model = VGG16(weights='imagenet')
```

```python
def prepare_and_predict(img_path):
    """ Förbereder bilden och gör förutsägelser. """
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    preds = model.predict(img_array)
    return decode_predictions(preds, top=3)[0]
```

```python
# Definiera sökvägar till dina bilder
image_paths = ['zebra.jpg', 'linjal.jpg', 'tennisboll.jpg']
```

```python
plt.figure(figsize=(12, 4))

for i, img_path in enumerate(image_paths):
    preds = prepare_and_predict(img_path)
    img_loaded = image.load_img(img_path, target_size=(224, 224))
    
    plt.subplot(1, 3, i + 1)
    plt.imshow(img_loaded)
    plt.title('\n'.join([f'{p[1]}: {p[2]:.2f}' for p in preds]))
    plt.axis('off')
```

```python
# Funktion för att förbereda bilder
def prepare_image(img_path, model):
    img = image.load_img(img_path, target_size=model.input_shape[1:3])
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return x
```

```python
# Funktion för att ladda modell och göra förutsägelser
def predict_image(model, x):
    preds = model.predict(x)
    return decode_predictions(preds, top=3)[0]
```

```python
# Ladda modellen
model = VGG16(weights='imagenet')
```

```python

image_paths = ['cyckel.jpeg', 'mugg.jpeg', 'stol.jpeg']

# Skapa en figur för att visa bilder och förutsägelser
plt.figure(figsize=(12, 4))
for i, img_path in enumerate(image_paths):
    x = prepare_image(img_path, model)
    preds = predict_image(model, x)
    img_loaded = image.load_img(img_path, target_size=(224, 224))
    
    plt.subplot(1, 3, i + 1)
    plt.imshow(img_loaded)
    plt.title('\n'.join([f'{p[1]}: {p[2]:.2f}' for p in preds]))
    plt.axis('off')

```

```python
from keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from keras.preprocessing import image
```

```python
def prepare_image(img_path, model):
    """Förbereder bilden för att kunna användas i modellen."""
    img = image.load_img(img_path, target_size=model.input_shape[1:3])
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return x
```

```python
def predict_image(model, x):
    """Gör förutsägelser för den givna bilden."""
    preds = model.predict(x)
    return decode_predictions(preds, top=3)[0]
```

```python
# Ladda MobileNetV2 modellen
model = MobileNetV2(weights='imagenet')
```

```python
image_paths = ['cyckel.jpeg', 'mugg.jpeg', 'stol.jpeg']
```

```python
plt.figure(figsize=(12, 4))
for i, img_path in enumerate(image_paths):
    x = prepare_image(img_path, model)
    preds = predict_image(model, x)
    img_loaded = image.load_img(img_path, target_size=(224, 224))
    
    plt.subplot(1, 3, i + 1)
    plt.imshow(img_loaded)
    plt.title('\n'.join([f'{p[1]}: {p[2]:.2f}' for p in preds]))
    plt.axis('off')
```

# 2. Answer the question: Why is "Pretrained models" a very powerful concept?
Pretrained models är ett kraftfullt koncept eftersom de utnyttjar redan existerande kunskap och parametrar från omfattande dataset. Detta möjliggör effektiv överföring av inlärning till nya uppgifter med minimala datakrav. Genom att använda dessa förtränade modeller kan man förbättra prestanda, spara tid och minska behovet av stora datamängder. Dessa modeller undviker även behovet av att skapa nya vikter, algoritmer och spridningstekniker från grunden, vilket leder till mer resurseffektiva lösningar.
